```python
def newton_iters(xn, num_iteraciones):
    iteraciones = [xn] # crea una lista
    errores = [abs(xn-1.)]
    for a in range(num_iteraciones):
        xn = xn - (xn*xn - 1.)/(2*xn)
        iteraciones.append(xn)
        errores.append(abs(xn-1))
    return iteraciones, errores

        
```

```python
it, er = newton_iters(2,5)
fig, axs = plt.subplots(1,2,tight_layout = True)
axs[0].plot(it, "x")
axs[0].set_title("Iteraciones")
axs[0].set_label("i")
axs[0].set_ylabel("xi")

axs[1].semilogy(er, "o")
axs[1].set_title("Error")
axs[1].set_label("i")
axs[1].set_ylabel("xi")
```

```python
#graficos en 3d
X = np.linspace(-2, 2)
Y = np.linspace(-1, 1)
x, y = np.meshgrid(X, Y) #llena el conjunto de valores con los valores de x y y
z = x**2 * y**3
fig = plt.figure()
ax = fig.add_subplot(projection="3d") 
ax.plot_surface(x, y, z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("Graph of the function f(x) = x^2y^3")
```
