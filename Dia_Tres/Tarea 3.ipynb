## Ejercicio 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5)
fig, ax = plt.subplots()
#3x + 2y = 6
l1 = ax.plot(x, ((6-3*x)/2))
#6x + 4y = -9
l2 = ax.plot(x, ((-9-6*x)/4))
#Leyendas
ax.set_title("Gráfico del 1er sistema de ecuaciones")
ax.legend(("3x + 2y = 6", "6x + 4y = -9"))
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.grid(True)
```

```python
x = np.linspace(0,5)
fig, ax = plt.subplots()
#x + y = 3
l1 = ax.plot(x, 3-x)
#x - 4y = 1
l2 = ax.plot(x, (-((1-x)/4)))
#Leyendas
ax.set_title("Gráfico del 2do sistema de ecuaciones")
ax.legend(("x + y = 3", "x - 4y = 1"))
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.grid(True)
```

```python
X = np.linspace(-5, 6)
Y = np.linspace(-5, 6)
x, y = np.meshgrid(X, Y)
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
#2x - y + 2z = 3
pl1 = ax.plot_surface(x,y,((3-2*x+y)/2))
#2x + 3y - z = 0
pl2 = ax.plot_surface(x,y,2*x+3*y)
#x + 2y + 2z = -12
pl3 = ax.plot_surface(x,y,(-((-12-x-2*y)/2)))
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Gráficos del 3er sistema de ecuaciones")
ax.grid(True)
```

## Ejercicio 2

```python
#primero definimos la funcion y = x*x + 3
def f(x):
    return x*x + 3
#punto que queremos obtener
x1 = 1
y1 = 4
#derivada de la funcion para obtener la pendiente de la recta
m = 2 * x1
x = np.linspace(-5,5)
tan = m * (x-x1) + y1
#realizamos las gráficas
fig, ax = plt.subplots()
l1 = ax.plot(x,f(x), color = 'blue')
l2 = ax.plot(x,tan,'--', color = 'red')
l3 = ax.plot(x1 , f(x1), 'o', color = 'red')
ax.set_title("Gráfico de la parábola y su recta tangente")
ax.legend(("Función", "Recta tangente","(1, 4)"))
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.grid(True)
```

## Ejercicio 3

```python
#usando el código del problema anterior, calculamos la recta tangente que pase por el punto (0.5,0)
def f(x):
    return x*x + 3

x1 = -1.3
y1 = 4.7

m = 2 * x1
x = np.linspace(-3,3)
tan = m * (x-x1) + y1

#punto por el que debe pasar la recta tangente
x2 = 0.5
y2 = 0

fig, ax = plt.subplots()
l1 = ax.plot(x,f(x), color = 'blue')
l2 = ax.plot(x,tan,'--', color ='green')
l3 = ax.plot(x1 , f(x1), 'o', color = 'green')
l4 = ax.plot(x2, y2, 'o', color = "red")
ax.set_title("Tiro al blanco")
ax.legend(("Función", "Recta tangente","(-1.3, 4.7)", "(0.5, 0)"))
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.grid(True)
```

## Ejercicio 4

```python
#definimos la funcion x*x*x - 2x*x - 2x
def f(x):
    return x**3 - x**2 - 2*x

a = 1
b = 3

#obtenemos la pendiente o la derivada de la funcion
m = (f(b) - f(a)) / (b - a)

x = np.linspace(-1,6)

#calculando la recta secante de la funcion
sec = f(a) + (m * (x - a))

#al igualar la pendiente con la derivada de la funcion obtenemos que x = 2.43
c = 2.43

#a partir de este resultado obtenemos la recta tangente
tan = m * (x - c) + f(c)

#graficamos
fig, ax = plt.subplots()
l1 = ax.plot(x,f(x), color = 'green')
l2 = ax.plot(x,sec, color = 'blue')
l3 = ax.plot(x,tan, color = 'red')
ax.set_title("Gráfico del teorema del valor medio")
ax.legend(("Función", "Recta secante","Recta tangente"))
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.grid(True)
```
